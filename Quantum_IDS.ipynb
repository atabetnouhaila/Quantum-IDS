```python
import numpy as np
import qiskit
from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.algorithms import QSVC
from qiskit.primitives import Estimator
from sklearn.model_selection import train_test_split
from qiskit.quantum_info import SparsePauliOp
from qiskit.circuit import QuantumCircuit

# Génération des données d'exemple
encoded_states = np.random.rand(100, 2)  # 100 échantillons avec 2 caractéristiques
labels = np.random.rand(100)

# Division en ensembles d'entraînement et de test
training_data, test_data, training_labels, test_labels = train_test_split(
    encoded_states, labels, test_size=0.2, random_state=42
)

# Création de la carte de caractéristiques
num_qubits = encoded_states.shape[1]
feature_map = ZZFeatureMap(feature_dimension=num_qubits, reps=2)

# Initialisation de l'estimator
estimator = Estimator()

def quantum_kernel(x1_vec, x2_vec):
    """
    Calcule le kernel quantique entre deux vecteurs de données.

    Args:
        x1_vec (np.ndarray): Premier vecteur de données
        x2_vec (np.ndarray): Second vecteur de données

    Returns:
        float: Valeur du kernel
    """
    # Création des circuits pour les deux vecteurs
    params = feature_map.parameters
    param_dict1 = dict(zip(params, x1_vec))
    param_dict2 = dict(zip(params, x2_vec))

    circuit1 = feature_map.assign_parameters(param_dict1)
    circuit2 = feature_map.assign_parameters(param_dict2)

    # Combinaison des circuits
    combined_circuit = QuantumCircuit(num_qubits)
    combined_circuit.compose(circuit1, inplace=True)
    combined_circuit.compose(circuit2.inverse(), inplace=True)

    # Mesure de l'état final
    observable = SparsePauliOp("Z" * num_qubits)
    job = estimator.run(combined_circuit, observable)
    result = job.result()
    return np.real(result.values[0])

# Test de la fonction kernel sur un seul exemple
print("Test du kernel sur un exemple...")
test_value = quantum_kernel(training_data[0], training_data[0])
print(f"Valeur du kernel pour le même point: {test_value}")

# Création de la matrice de kernel pour l'ensemble d'entraînement
print("\nCalcul de la matrice de kernel pour l'ensemble d'entraînement...")
kernel_matrix = np.zeros((len(training_data), len(training_data)))
for i in range(len(training_data)):
    for j in range(len(training_data)):
        kernel_matrix[i, j] = quantum_kernel(training_data[i], training_data[j])
    if i % 10 == 0:  # Affichage de la progression
        print(f"Progression: {i}/{len(training_data)} échantillons traités")

# Vérification que la matrice de kernel est semi-définie positive
eigenvals = np.linalg.eigvals(kernel_matrix)
min_eigenval = np.min(np.real(eigenvals))
print(f"\nPlus petite valeur propre de la matrice de kernel: {min_eigenval}")
if min_eigenval < -1e-10:  # permet une petite erreur numérique
    print("Attention: La matrice de kernel n'est pas semi-définie positive!")
    # Correction numérique si nécessaire
    kernel_matrix -= min_eigenval * np.eye(len(kernel_matrix))

# Initialisation et entraînement du modèle SVR
from sklearn.svm import SVR
print("\nEntraînement du modèle SVR...")
qsvr = SVR(kernel='precomputed')
qsvr.fit(kernel_matrix, training_labels)

# Préparation de la matrice de kernel pour les prédictions
print("\nCalcul de la matrice de kernel pour l'ensemble de test...")
test_kernel = np.zeros((len(test_data), len(training_data)))
for i in range(len(test_data)):
    for j in range(len(training_data)):
        test_kernel[i, j] = quantum_kernel(test_data[i], training_data[j])
    if i % 5 == 0:  # Affichage de la progression
        print(f"Progression: {i}/{len(test_data)} échantillons traités")

# Prédictions
print("\nGénération des prédictions...")
predictions = qsvr.predict(test_kernel)

# Évaluation du modèle
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(test_labels, predictions)
r2 = r2_score(test_labels, predictions)

print("\nRésultats:")
print("Erreur quadratique moyenne:", mse)
print("Score R2:", r2)
print("\nQuelques prédictions:")
print("Vraies valeurs:", test_labels[:5])
print("Prédictions:", predictions[:5])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, auc, det_curve
from scipy.stats import norm
import seaborn as sns

def plot_performance_metrics(y_true, y_pred, y_scores=None):
    """
    Crée un ensemble complet de visualisations pour l'évaluation du modèle.

    Args:
        y_true: Les vraies étiquettes
        y_pred: Les prédictions du modèle
        y_scores: Les scores de probabilité (si disponibles)
    """
    # Convertir les valeurs continues en classes binaires pour l'analyse
    threshold = 0.5
    y_true_binary = (y_true > threshold).astype(int)
    y_pred_binary = (y_pred > threshold).astype(int)

    # Créer une figure avec plusieurs sous-graphiques
    fig = plt.figure(figsize=(20, 15))

    # 1. Matrice de confusion
    plt.subplot(2, 2, 1)
    cm = confusion_matrix(y_true_binary, y_pred_binary)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Matrice de Confusion')
    plt.xlabel('Prédictions')
    plt.ylabel('Vraies Valeurs')

    # 2. Courbes FAR/FRR
    plt.subplot(2, 2, 2)
    thresholds = np.linspace(0, 1, 100)
    far = []
    frr = []

    for thresh in thresholds:
        y_pred_th = (y_pred > thresh).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true_binary, y_pred_th).ravel()

        far.append(fp / (fp + tn))  # False Acceptance Rate
        frr.append(fn / (fn + tp))  # False Rejection Rate

    plt.plot(thresholds, far, label='FAR')
    plt.plot(thresholds, frr, label='FRR')
    plt.xlabel('Seuil')
    plt.ylabel('Taux')
    plt.title('Courbes FAR/FRR')
    plt.legend()
    plt.grid(True)

    # 3. Courbe DET
    plt.subplot(2, 2, 3)
    if y_scores is None:
        y_scores = y_pred

    fpr, fnr, _ = det_curve(y_true_binary, y_scores)

    # Conversion en échelle normale
    plt.plot(norm.ppf(fpr), norm.ppf(fnr))
    plt.xlabel('False Positive Rate (échelle normale)')
    plt.ylabel('False Negative Rate (échelle normale)')
    plt.title('Courbe DET')
    plt.grid(True)

    # 4. Distribution des prédictions
    plt.subplot(2, 2, 4)
    plt.hist(y_pred[y_true_binary == 0], bins=20, alpha=0.5, label='Classe 0', density=True)
    plt.hist(y_pred[y_true_binary == 1], bins=20, alpha=0.5, label='Classe 1', density=True)
    plt.xlabel('Score de prédiction')
    plt.ylabel('Densité')
    plt.title('Distribution des Prédictions par Classe')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Afficher quelques métriques supplémentaires
    tn, fp, fn, tp = confusion_matrix(y_true_binary, y_pred_binary).ravel()
    print("\nMétriques de performance:")
    print(f"Accuracy: {(tp + tn) / (tp + tn + fp + fn):.3f}")
    print(f"Taux de faux positifs (FAR): {fp / (fp + tn):.3f}")
    print(f"Taux de faux négatifs (FRR): {fn / (fn + tp):.3f}")
    print(f"EER: {np.interp(0.5, far, frr):.3f}")

# Utilisation de la fonction
plot_performance_metrics(test_labels, predictions)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import qiskit
from qiskit.circuit.library import ZZFeatureMap
from qiskit.primitives import Estimator
from qiskit.quantum_info import SparsePauliOp
from qiskit.circuit import QuantumCircuit
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

def prepare_data(df, n_components=4):
    """
    Prépare les données pour l'apprentissage quantique.
    """
    X = df.drop(['category', 'proto', 'Unnamed: 0'], axis=1)
    y = df['category']

    le = LabelEncoder()
    y = le.fit_transform(y)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    pca = PCA(n_components=n_components)
    X_reduced = pca.fit_transform(X_scaled)

    print(f"Variance expliquée par les {n_components} composantes: {pca.explained_variance_ratio_.sum():.3f}")

    return X_reduced, y, le.classes_, pca, scaler

def quantum_kernel(x1_vec, x2_vec, feature_map, estimator):
    """
    Calcule le kernel quantique entre deux vecteurs de données.
    """
    print(f"Calcul du kernel quantique pour les vecteurs: {x1_vec} et {x2_vec}")

    params = feature_map.parameters
    param_dict1 = dict(zip(params, x1_vec))
    param_dict2 = dict(zip(params, x2_vec))

    circuit1 = feature_map.assign_parameters(param_dict1)
    circuit2 = feature_map.assign_parameters(param_dict2)

    combined_circuit = QuantumCircuit(len(x1_vec))
    combined_circuit.compose(circuit1, inplace=True)
    combined_circuit.compose(circuit2.inverse(), inplace=True)

    observable = SparsePauliOp("Z" * len(x1_vec))
    job = estimator.run(combined_circuit, observable)
    result = job.result()

    print(f"Résultat du job quantique: {result.values[0]}")
    return np.real(result.values[0])

def train_quantum_svm(X_train, X_test, y_train, y_test, n_qubits):
    """
    Entraîne le SVM quantique et évalue ses performances.
    """
    feature_map = ZZFeatureMap(feature_dimension=n_qubits, reps=2)
    estimator = Estimator()

    print("Calcul de la matrice de kernel pour l'ensemble d'entraînement...")
    kernel_matrix = np.zeros((len(X_train), len(X_train)))
    for i in range(len(X_train)):
        for j in range(len(X_train)):
            kernel_matrix[i, j] = quantum_kernel(X_train[i], X_train[j], feature_map, estimator)
        if i % 10 == 0:
            print(f"Progression: {i}/{len(X_train)}")

    print("Entraînement du SVM...")
    qsvm = SVC(kernel='precomputed')
    qsvm.fit(kernel_matrix, y_train)

    print("Calcul de la matrice de kernel pour l'ensemble de test...")
    test_kernel = np.zeros((len(X_test), len(X_train)))
    for i in range(len(X_test)):
        for j in range(len(X_train)):
            test_kernel[i, j] = quantum_kernel(X_test[i], X_train[j], feature_map, estimator)
        if i % 5 == 0:
            print(f"Progression: {i}/{len(X_test)}")

    predictions = qsvm.predict(test_kernel)
    return predictions

def main(df, n_components=4):
    """
    Fonction principale d'exécution avec vérification des classes.
    """
    if len(df['category'].unique()) < 2:
        raise ValueError("Le dataset doit contenir au moins 2 classes différentes")

    X_reduced, y, classes, pca, scaler = prepare_data(df, n_components)

    X_train, X_test, y_train, y_test = train_test_split(
        X_reduced, y, test_size=0.2, random_state=42,
        stratify=y
    )

    print("Entraînement du modèle et génération des prédictions...")
    predictions = train_quantum_svm(X_train, X_test, y_train, y_test, n_components)

    print("\nRapport de classification:")
    print(classification_report(y_test, predictions, target_names=classes))

    print("\nMatrice de confusion:")
    print(confusion_matrix(y_test, predictions))

    return pca, scaler

if __name__ == "__main__":
    # Chargement et préparation des données
    data = pd.read_csv('/content/train1.csv')


    # Échantillonnage équilibré
    n_samples_per_class = 100
    df_sampled = pd.DataFrame()

    for category in data['category'].unique():
        category_data = data[data['category'] == category].head(n_samples_per_class)
        df_sampled = pd.concat([df_sampled, category_data])

    df_sampled = df_sampled.reset_index(drop=True)
    pca, scaler = main(df_sampled, n_components=4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import norm

def plot_regression_metrics(y_true, y_pred):
    """
    Crée un ensemble de visualisations pour l'évaluation du modèle de régression.

    Args:
        y_true: Les vraies valeurs
        y_pred: Les prédictions du modèle
    """
    # Créer une figure avec plusieurs sous-graphiques
    fig = plt.figure(figsize=(20, 15))

    # 1. Scatter plot des prédictions vs vraies valeurs
    plt.subplot(2, 2, 1)
    plt.scatter(y_true, y_pred, alpha=0.5)
    min_val = min(y_true.min(), y_pred.min())
    max_val = max(y_true.max(), y_pred.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ligne parfaite')
    plt.xlabel('Vraies valeurs')
    plt.ylabel('Prédictions')
    plt.title('Prédictions vs Vraies valeurs')
    plt.legend()
    plt.grid(True)

    # 2. Distribution des erreurs
    plt.subplot(2, 2, 2)
    errors = y_pred - y_true
    sns.histplot(errors, kde=True)
    plt.xlabel('Erreur de prédiction')
    plt.ylabel('Fréquence')
    plt.title('Distribution des erreurs')
    plt.grid(True)

    # 3. Résidus vs Prédictions
    plt.subplot(2, 2, 3)
    plt.scatter(y_pred, errors, alpha=0.5)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel('Prédictions')
    plt.ylabel('Résidus')
    plt.title('Résidus vs Prédictions')
    plt.grid(True)

    # 4. Q-Q plot des résidus
    plt.subplot(2, 2, 4)
    from scipy.stats import probplot
    probplot(errors, dist="norm", plot=plt)
    plt.title('Q-Q Plot des résidus')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Afficher les métriques de performance
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    mae = np.mean(np.abs(errors))

    print("\nMétriques de performance:")
    print(f"MSE (Mean Squared Error): {mse:.6f}")
    print(f"RMSE (Root Mean Squared Error): {rmse:.6f}")
    print(f"MAE (Mean Absolute Error): {mae:.6f}")
    print(f"R² Score: {r2:.6f}")
    print(f"Erreur moyenne: {np.mean(errors):.6f}")
    print(f"Écart-type des erreurs: {np.std(errors):.6f}")

# Utiliser la fonction après avoir exécuté le modèle
plot_regression_metrics(test_labels, predictions)
```
